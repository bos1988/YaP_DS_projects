### Подключаемые модули, общие функции: <a id='functions'></a>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as st

from pymystem3 import Mystem
from collections import Counter

import warnings
warnings.simplefilter('ignore')
```

```python
# ПАРАМЕТРЫ
table_size = 100

# ФУНКЦИИ:

# Вывод информации о всех таблицах:
def display_all(df, names, functions):
    size = table_size
    for name in names: 
        print('_'*size)
        print()
        table_name = ' ТАБЛИЦА ' + name + ' '
        print(table_name + table_name.rjust(size-len(table_name), '='))
        if type(functions) == str:
            display(eval(f"df['{name}'].{functions}"))
        else:
            for func in functions:
                print('.'*size)
                print(func.rjust(size))
                display(eval(f"df['{name}'].{func}"))
        print()
        
def print_descripltion(df, description):
    limit_text = table_size - 18
    for index in df:
        end = '' if len(description[index][1]) < limit_text else '...'
        print(f'{index:<18} - {description[index][1][:limit_text]}{end}')

def get_shape(data):
    shape = {}
    for name in data.keys():
        shape[name] = data[name].shape
    shape = pd.DataFrame(shape).T
    shape.set_axis(['rows','cols'], axis=1, inplace=True)
    return shape

# Рассчет номера месяца пользования услугами
def set_use_month(date_0, date):
    return (date.year - date_0.year)*12 + date.month - date_0.month + (date.day >= date_0.day)

#=====================================================================================================================
# Презентации:
def print_plot(data, title=None, xlabel=None, ylabel=None, s_grid=True, figsize=(10, 6), **kwargs):
    fig, ax = plt.subplots()
    data.plot(ax=ax, **kwargs)
    ax.minorticks_on()
    if s_grid:
        ax.grid(which='major',
                color = 'grey', 
                linewidth = 1.2)
        ax.grid(which='minor', 
                color = 'grey', 
                linestyle = ':')
    fig.set_figwidth(figsize[0])
    fig.set_figheight(figsize[1])
    ax.tick_params(labelsize = 12)
    ax.set_title(title, pad=15, fontsize = 16)
    ax.set_xlabel(xlabel, fontsize = 14)
    ax.set_ylabel(ylabel, fontsize = 14)

def hist_describe(table, col, title=None, xlabel=None, ylabel=None, data=None, **kwargs):
    if type(data) == type(None):
        data = df[table][col]
    name = description[table][col][0]
    if xlabel == None:
        xlabel = name
    print_plot(data, f'Распределение показателя "{name}"',
               xlabel, 'Частота', s_grid=False, kind='hist', grid=True, **kwargs)
    display(data.describe())

def box_describe(table, col, title=None, xlabel=None, ylabel=None, data=None, **kwargs):
    if type(data) == type(None):
        data = df[table][col]
    name = description[table][col][0]
    if xlabel == None:
        xlabel = name
    print_plot(data, f'Диаграмма размаха для показателя "{name}"',
               xlabel, ylabel, s_grid=False, kind='box', figsize=(10, 3),
               vert=False, widths=0.8, grid=True, **kwargs)
    display(data.describe())

def hist_box_describe(table, col, title=None, xlabel=None, ylabel=None, data=None, bins=None, **kwargs):
    if type(data) == type(None):
        data = df[table][col]
    try:
        name = description[table][col][0]
    except:
        pass
    if title != None:
        name = title
    if xlabel == None:
        xlabel = name
    print_plot(data, f'Распределение показателя "{name}"',
               xlabel, 'Частота', s_grid=False, kind='hist', grid=True, bins=bins, **kwargs)
    print_plot(data, f'Диаграмма размаха для показателя "{name}"',
               xlabel, ylabel, s_grid=False, kind='box', figsize=(10, 3),
               vert=False, widths=0.8, grid=True, **kwargs)
    display(data.describe())
```

Просчитаем количество значений выбивающихся из этого правила:

### Помесячная выручка с каждого пользователя

Создадим общую сводную таблицу по объемам потребления услуг. Внесем туда:
- суммарное время звонков за месяц
- суммарное время звонков за месяц в минутах
- количество звонков в месяц
- суммарный объем интернет-трафика за месяц (в Mb)
- суммарный объем интернет-трафика за месяц (в Gb)
- суммарное количество сообщений за месяц
- наименование тарифа
- наименование населенного пункта

```python
# ['calls_month', 'internet_month', 'n_messages_month']
def create_summary_table(data):
    return (data['calls_month']
            .merge(data['internet_month'], on=['user_id','month'], how='outer')
            .merge(data['n_messages_month'], on=['user_id','month'], how='outer')
            .merge(data['users'][['user_id','tariff', 'city']], on='user_id', how='right')
            )
# merge USERS how=right, поскольку есть пользователи не использовавшие услуги, но платившие абонентскую плату

df_summary = create_summary_table(df)
df_summary_treated = create_summary_table(df_treated)
```

```python
df_summary.head()
```

Рассчитаем выручку по месяцам

*Информация о тарифах:*

```python
df['tariffs']
```

Расчет:

```python
# Словарь параметров тарифов дял функции расчета
tariff_parameters = {}
for i, tariff in df['tariffs'].iterrows():
    tariff_parameters[tariff['tariff_name']] = dict(tariff)

# функция рассчета
def calc_payment(row):
    tarif_name = row['tariff']
    tariff = tariff_parameters[tarif_name]
    result = tariff['rub_monthly_fee']
    
    minutes = row['minutes'] - tariff['minutes_included']
    gb = row['gb_used'] - tariff['mb_per_month_included'] / 1024
    messages = row['count_message'] - tariff['messages_included']
    
    minutes = minutes if minutes > 0 else 0
    gb = gb if gb > 0 else 0
    messages = messages if messages > 0 else 0
    
    result += minutes * tariff['rub_per_minute'] + gb * tariff['rub_per_gb'] + messages * tariff['rub_per_message']
    return result
```

```python
df_summary['payment'] = df_summary.apply(calc_payment, axis=1)
df_summary_treated['payment'] = df_summary_treated.apply(calc_payment, axis=1)
```

```python
def analysis(data, col, param):
    indent = 26
    roundoff = 3
    hist_box_describe(None, None , data=data[col], bins=100, title=param)
    print('_'*table_size)
    print()
    print(f'КЛЮЧЕВЫЕ МЕТРИКИ по показателю "{param}":')
    print('_'*table_size)
    print()
    mean_temp = data[col].mean()
    std_temp = np.std(data[col], ddof=1)
    print('Медианное значение:'.ljust(indent), round(data[col].median(), roundoff))
    print('Среднее значение:'.ljust(indent), round(mean_temp, roundoff))
    print('Несмещенная оценка дисперсиии по выборке:'.ljust(indent), round(np.var(data[col], ddof=1), roundoff))
    print('Оценка стандартного отклонения:'.ljust(indent),round(std_temp, roundoff))
    print('_'*table_size)
    print()
    plt.show()
    print('Редкие и выбивающиеся значения:')
    rare = data.query(f'{col} > {mean_temp + std_temp * 3} or {col} < {mean_temp - std_temp * 3}')
    if len(rare) > 6:
        print(f'Выбивающихся значений - {len(rare)} позиций')
    else:
        display(rare)
    print()
```
