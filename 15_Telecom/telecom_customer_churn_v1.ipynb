# Прогнозирование оттока клиентов телеком компании

Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

### Описание услуг

Оператор предоставляет два основных типа услуг: 

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL*,* от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*).  

Также доступны такие услуги:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

### Описание данных

Данные состоят из файлов, полученных из разных источников:

- `contract.csv` — информация о договоре;
- `personal.csv` — персональные данные клиента;
- `internet.csv` — информация об интернет-услугах;
- `phone.csv` — информация об услугах телефонии.

Во всех файлах столбец `customerID` содержит код клиента.

Информация о договорах актуальна на 1 февраля 2020.

```python
def camel_to_snake(string, sep='_'):
    camel = ''
    for i in range(len(string)):
        camel += sep if string[i].isupper() and string[i-1].islower() else ''
        camel += string[i].lower()
    return camel.lstrip(sep)
```

```python
# перевод из верблюжъего в змеиный регистр для словаря описания данных
def description_camel_to_snake(description_dict):
    camel_description = dict()
    for key, val in description_dict.items():
        camel_description[camel_to_snake(key)] = val

    return camel_description
```

```python
# разбивка на валидационную и тестовую выборки
def research_test_split(df, test_size=0.2, random_state=RANDOM_STATE):
    df_research, df_test = train_test_split(df, test_size=test_size, random_state=random_state)
    print()
    display(pd.DataFrame({
        'Выборка:': ['Исследовательская', 'Тестовая'],
        'Количество значений:': [df_research.shape[0], df_test.shape[0]],
        'Доля выборки:': [df_research.shape[0] / df.shape[0], df_test.shape[0] / df.shape[0]]
            }).style.format({'Количество значений': '{:.0f}', 'Доля выборки:': '{:.2%}'})
         )
    return df_research, df_test
```

```python
# Рабочая версия с обработкой Pipeline
# Проходит не по всей сетке, а пока растет значение метрики (Бейесовская оптимизация).
# Если метрика не растет на протажении хвоста (max_tail) цикл заканчивается
# На вход подается Класс модели, исследуемые гиперапараметры в виде словаря
# Работает быстрее, чем GridSearchCV
def ascending_grid(df_features, df_target, func, parameters, scoring=None, plot=False, subtitle='', max_tail=None, cv=5, n_jobs=-1, random_state=RANDOM_STATE, **kwargs):
    #cv = 5 #при значении 5 результаты отличаются от результатов при значении > 10 ! но по умолчанию в cross_val_score стоит 5, поэтому оставим по умоланию 5
    steps = 100
    iterator = 0
    best_score = -1_000_000
    best_param = parameters
    tail = 0
    if max_tail == None:
        max_tail = 1 if 5 - len(parameters) < 1 else 5 - len(parameters)
    
    param = list(parameters.keys())[0]
    start = parameters[param][0]
    sep = parameters[param][1]
    
    if len(parameters) == 1:
        for p in np.arange(start, start+steps*sep, sep):
            set_param = {param: p}
            if type(func) == Pipeline:
                model = func
                model_name = model.get_params().get("steps")[-1][0]
                # Гиперпараметры (стандартные random_state=123, n_jobs=n_jobs задаются извне):
                # Текущая настройка
                model.set_params(**{model_name+'__'+param: p})
                # **kwargs
                params_temp = {}
                for key, val in kwargs.items():
                    params_temp[model_name+'__'+key] = val
                model.set_params(**params_temp)
            else:
                model = func(random_state=random_state, n_jobs=n_jobs, **set_param, **kwargs)
            scores = cross_val_score(model, df_features, df_target, scoring=scoring, cv=cv)
            score = scores.mean()
            iterator += len(scores)
            if plot:
                print(f'{p} >>> {set_param} {kwargs} - {score:.4f}')

            if score > best_score:
                best_score = score
                set_param.update(kwargs)
                best_param = set_param
                tail = 0
            else:
                tail += 1
            if tail >= max_tail:
                break
    else:
        # рекурсия
        parameters_minus = parameters.copy()
        del parameters_minus[param]
        for p in np.arange(start, start+steps*sep, sep):
            set_param = {param: p}
            result = ascending_grid(df_features, df_target, func, parameters_minus,
                                    scoring=scoring, plot=plot, subtitle=subtitle, max_tail=max_tail, cv=cv,
                                    **set_param, **kwargs)
            iterator += result['iterations']
            score = result['score']
            
            # Сокращение нижней границы поиска (предыдущее лучшее минус 2 промежутка)
            for key in result['parameters'].keys():
                try:
                    new_start = result['parameters'][key] - parameters_minus[key][1]*2
                    if parameters_minus[key][0] < new_start:
                        parameters_minus[key][0] = result['parameters'][key] - parameters_minus[key][1] * (max_tail + 1)
                except:
                    pass
            
            if score > best_score:
                best_score = score
                best_param = result['parameters']
                tail = 0
            else:
                tail += 1
            if tail >= max_tail:
                break
        
        if plot:
            print('best_score =', best_score)
            print('best_param =', best_param)
    if plot:
        print('-'*TABLE_SIZE)
    return ({'iterations': iterator, 'score':best_score, 'parameters':best_param})
```

```python
#gscv.cv_results_
def grid_search_cv(X, y, model, parameters, scoring=None, cv=5, plot=False):
    verbose = (0, 3)[plot]
    gscv = GridSearchCV(model, parameters, scoring=scoring, cv=cv, n_jobs=-1, verbose=verbose)
    gscv.fit(X, y)
    results = {
        'iterations': len(gscv.cv_results_['params']),
        'score': gscv.best_score_,
        'parameters': gscv.best_params_
    }
    return results
```

```python
# Кросс-валидация с результатом для одной модели
def cross_val_base(model_name, model, X, y, scoring, print_res=True):
    time_start = time.time()
    scores = cross_val_score(model, X, y, scoring=scoring, cv=10)
    if print_res:
        print(f'Среднее значение метрики {scoring} для базовой модели {text.BOLD}{model_name}:\n{text.BG_SILVER} {scoring} = {scores.mean():.2f} {text.END}')
        print('-'*TABLE_SIZE)
        return
    return (model_name, scores.mean(), time.time() - time_start)
```

```python
# вывод таблицы сравнения
def get_selection_model_results(results, score_name = 'Score'):
    res = pd.DataFrame(results, columns = ['Model', score_name, 'Time']).sort_values(by=score_name, ascending=False)
    
    res['rating_score'] = res.apply(
        lambda row: round((row[score_name] - res[score_name].min()) / (res[score_name].max() - res[score_name].min()), 3)
        , axis=1)
    res['rating_time'] = res.apply(
        lambda row: round(-(row['Time'] - res['Time'].min()) / (res['Time'].max() - res['Time'].min()), 3)
        , axis=1)
    res['rating_total'] = res['rating_score'] + res['rating_time']
    
    return res
```

```python
# вывод итоговых метрик
def get_metrics(model, train_XY:tuple, test_XY:tuple):
    metrics = []

    probabilities_train = model.predict_proba(train_XY[0])[:, 1]
    probabilities_test = model.predict_proba(test_XY[0])[:, 1]
    
    predictions_train = model.predict(train_XY[0])
    predictions_test = model.predict(test_XY[0])

    metrics.append(['roc_auc', roc_auc_score(train_XY[1], probabilities_train), roc_auc_score(test_XY[1], probabilities_test)])
    metrics.append(['f1', f1_score(train_XY[1], predictions_train), f1_score(test_XY[1], predictions_test)])
    metrics.append(['accuracy', accuracy_score(train_XY[1], predictions_train), accuracy_score(test_XY[1], predictions_test)])
    
    return pd.DataFrame(metrics, columns = ['Метрика', 'Обучающая', 'Тестовая']).set_index('Метрика')
```

```python
# вывод итоговых метрик
def get_metrics_ensemble(proda_train, proba_test, Y_train, Y_test):
    metrics = []
    
    predictions_train = proda_train > 0.5
    predictions_test = proba_test > 0.5

    metrics.append(['roc_auc', roc_auc_score(Y_train, proda_train), roc_auc_score(Y_test, proba_test)])
    metrics.append(['f1', f1_score(Y_train, predictions_train), f1_score(Y_test, predictions_test)])
    metrics.append(['accuracy', accuracy_score(Y_train, predictions_train), accuracy_score(Y_test, predictions_test)])
    
    return pd.DataFrame(metrics, columns = ['Метрика', 'Обучающая', 'Тестовая']).set_index('Метрика')
```

**2. lifetime**

Если же исключить признак *точки отсчета* `begin_date`, то время жизни договора измеренное в днях также будет являться утечкой целевого признака. Поскольку все даты в наборе данных имеют только даты на 1-е число, то возникает влияение количества дней в определнных месяцах отсчета.

Рассмотрим значения время жизни договора для обоих значений целевого признака:

```python
df_periods = pd.DataFrame(df_clean.query('outflow == 1')['lifetime_d'].value_counts().rename('Целевой 1')).join(
    df_clean.query('outflow == 0')['lifetime_d'].value_counts().rename('Целевой 0')
).sort_index()

df_periods.head(15)
```

Из-за того, что в определенных месяцах определенное кол-во дней, периоды жизни договора могут сигнализировать о возможности окончания договора 01.02.2020, т.е. модель может обучиться на конкрентную дату, чего нельзя допустить.

Например, если время жизни договора - 61 день, то это не может быть 01.12.2019, поскольку в декабре и январе по 31 дню - в сумме 62. Модель выучит, что при времени жизни договора - 61 день - это ушедший клиент. При этом происходит вляние даты 01.02.2020 на предсказания, т.е. просачивается значение из `EndDate` (истинного целевого признака).

Это тоже самое, что пометить (*закодировать*) целевые положительные метки определенными цифрами.

Модель выучит эту кодировку, подстроится под дату выгрузки (01.02.2020) и при изменении даты выгрузки ошибка модели сильно возрастет.

```python
print(f"Доля утечки целевого признака: {df_periods[df_periods['Целевой 0'].isna()]['Целевой 1'].sum() / df_periods['Целевой 1'].sum():.2%}")
```

```python
def get_research_test_m_ohe(df, features):
    df_major = make_df_major(df, features)
    df_major_ohe = pd.get_dummies(df_major, drop_first=True)
    
    # возвращаем df_research, df_test
    return research_test_split(df_major_ohe)
```

#### Порядковое кодирование

Применим технику порядкового кодирования (Ordinal Encoding, OE) для преобразования всех категориальных признаков в численные.

```python
# на вход подается pandas.DataFrame с признаками которые нужно кодировать типа 'category'
def get_research_test_m_oe(df, features):
    df_major = make_df_major(df, features)
    cat_feat = df_major[make_column_selector(dtype_include='category')].columns
    
    # через ColumnTransformer получается array без подписей с перемешанными колонками, поэтому так удобнее:
    encoder = OrdinalEncoder()
    df_major_oe = df_major.copy().reset_index(drop=True)
    df_major_oe[cat_feat] = encoder.fit_transform(df_major[cat_feat])
    df_major_oe[cat_feat] = df_major_oe[cat_feat].astype('category')
    
    # возвращаем df_research, df_test
    return research_test_split(df_major_oe)
```

Установим искомую метрику

> Ошибки классификации бывают двух видов: False Positive и False Negative. В статистике первый вид ошибок называют ошибкой I-го рода, а второй — ошибкой II-го рода. В нашей задаче по определению оттока абонентов, ошибкой первого рода будет принятие лояльного абонента за уходящего, так как наша нулевая гипотеза состоит в том, что никто из абонентов не уходит, а мы эту гипотезу отвергаем. Соответственно, ошибкой второго рода будет являться "пропуск" уходящего абонента и ошибочное принятие нулевой гипотезы.
<br/><br/>
В данной задаче нужно упустить как можно меньше клиентов, которые могут уйти (ошибка II-го рода). И желательно не раздавать промокоды и специалные условия клиентам, которые уходить не собираются (ошибка I-го рода).
<br/><br/>
За ошибку II-го рода может отвечать True Positive Rate (TPR) (или полнота recall), определяющая какую долю уходящих клиентов из всех действительно уходящих нашел алгоритм.
<br/><br/>
За ошибку I-го рода может отвечать False Positive Rate (FPR), определяющая какую долю из остающихся клиентов алгоритм предсказал неверно.
<br/><br/>
Чтобы минимизировать обе ошибки нужно получить как можно лучше метрику **ROC AUC**, учитывающую соотношение двух показателей TPR и FPR.

Функция оценки важности признаков

```python
# Оценка важности признаков
def get_important_columns(model, X, y, scoring, n_repeats=30, random_state=RANDOM_STATE):
    model.fit(X, y)
    r = permutation_importance(model, X, y,
                               scoring = scoring,
                               n_repeats=n_repeats,
                               random_state=random_state)
    for i in r.importances_mean.argsort()[::-1]:
        if r.importances_mean[i] - 2 * r.importances_std[i] > 0:
            print(f"{X.columns[i]:<40}"
                  f"{r.importances_mean[i]:.6f}"
                  f" +/- {r.importances_std[i]:.6f}")
    return X.columns[r.importances_mean.argsort()[::-1]]
```

## Рабочий код

Функции обработки:

```python
class WorkModel():
    def __init__(self):
        std_param = {'random_state': RANDOM_STATE, 'n_jobs': -1, 'class_weight': 'balanced'}
        
        self.features_important_booster = [
            'Type', 'PaperlessBilling', 'PaymentMethod',
            'MonthlyCharges', 'TotalCharges',
            'gender', 'SeniorCitizen',
            'OnlineSecurity', 'OnlineBackup', 'TechSupport', 'StreamingTV', 'StreamingMovies',
            'InternetService', 'MultipleLines',
            'n_pays',
            ]
        self.features_important_forest = [
            'Type', 'PaperlessBilling', 'PaymentMethod',
            'MonthlyCharges', 'TotalCharges',
            'gender', 'SeniorCitizen', 'Partner',
            'OnlineSecurity', 'OnlineBackup', 'TechSupport', 'StreamingTV', 'StreamingMovies',
            'InternetService', 'MultipleLines',
            'n_pays',
            ]
        self.features_important_linear = [
            'Type', 'PaperlessBilling', 'PaymentMethod',
            'MonthlyCharges', 'TotalCharges',
            'SeniorCitizen',
            'OnlineSecurity', 'TechSupport', 'StreamingTV', 'StreamingMovies',
            'InternetService', 'MultipleLines',
            'n_pays',
            ]

        best_LGBMClassifier = {
            'learning_rate': 0.24,
            'n_estimators': 160,
            'num_leaves': 2,
            'max_depth': 2}
        best_RandomForestClassifier = {
            'max_depth': 8,
            'n_estimators': 130,
            'min_samples_leaf': 3,
            'min_samples_split': 3}
        best_LogisticRegression = {'C': 150.0, 'penalty': 'l2'}
        
        tr_oe = ColumnTransformer([
                ("oe", OrdinalEncoder(), make_column_selector(dtype_include='category')),
            ],
            remainder="passthrough"
        )
        tr_ohe = ColumnTransformer([
                ("ohe", OneHotEncoder(), make_column_selector(dtype_include='category')),
            ],
            remainder="passthrough"
        )
        
        self.model_lgbm = LGBMClassifier(**best_LGBMClassifier, **std_param)
        self.model_forest = Pipeline([
            ("oe", tr_oe),
            ("model", RandomForestClassifier(**best_RandomForestClassifier, **std_param))
        ])
        
        self.model_linear = Pipeline([
            ("ohe", tr_ohe),
            ("scale", StandardScaler()),
            ("model", LogisticRegression(**best_LogisticRegression, **std_param))
        ])
    
    def train_preparation(self, df_contract, df_personal, df_internet, df_phone):
        features = [
            # 'BeginDate',
            'Type', 'PaperlessBilling', 'PaymentMethod',
            'MonthlyCharges', 'TotalCharges',
            'gender', 'SeniorCitizen', 'Partner', 'Dependents',
            'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
            'InternetService', 'phone', 'MultipleLines',
            'n_pays',
            # 'EndDate',
        ]
        cat_features = ['Type', 'PaymentMethod', 'gender', 'InternetService']
        
        # Обработка признаков
        # df_contract['BeginDate'] = pd.to_datetime(df_contract['BeginDate'], format='%Y-%m-%d')
        df_contract['TotalCharges'] = df_contract['TotalCharges'].replace(r'\s+', 0, regex=True).astype('float')
        df_contract['n_pays'] = round(df_contract['TotalCharges'] / df_contract['MonthlyCharges'], 0).astype('int')
        df_phone['phone'] = 1
        
        # Объединение таблиц
        df_all = df_contract\
            .merge(df_personal, 
                on = 'customerID',
                how = 'outer')\
                    .merge(df_internet, 
                        on = 'customerID',
                        how = 'outer')\
                            .merge(df_phone, 
                                on = 'customerID',
                                how = 'outer')
        
        # Обработка признаков
        df_all = df_all[features].fillna('No')
        df_all['InternetService'] = df_all['InternetService'].replace(0, 'None')
        df_all[cat_features] = df_all[cat_features].astype('category')
#         df_all['BeginDate'] = df_all['BeginDate'].dt.month
            
        for col in df_all[make_column_selector(dtype_include='object')].columns:
            df_all[col] = df_all[col].apply(lambda x: 0 if x == 'No' else 1)
        
        return df_all   
    
    def fit(self, df_contract, df_personal, df_internet, df_phone):
        
        df_contract, y_train = xy_split(df_contract, 'EndDate')
        y_train = y_train.apply(lambda x: 0 if x == 'No' else 1)
        
        X_train = self.train_preparation(df_contract, df_personal, df_internet, df_phone)
        
        self.model_lgbm.fit(X_train[self.features_important_booster], y_train)
        self.model_forest.fit(X_train[self.features_important_forest], y_train)
        self.model_linear.fit(X_train[self.features_important_linear], y_train)
#         self.model_linear.fit(pd.get_dummies(X_train, drop_first=True)[self.features_important_linear], y_train)
# get_dummies не проходит на тесте - нужен Pipeline
        
    def predict_proba(self, df_contract, df_personal, df_internet, df_phone):
        
        X_test = self.train_preparation(df_contract, df_personal, df_internet, df_phone)
        
        prob_train_1 = self.model_lgbm.predict_proba(X_test[self.features_important_booster])[:, 1]
        prob_train_2 = self.model_forest.predict_proba(X_test[self.features_important_forest])[:, 1]
        prob_train_3 = self.model_linear.predict_proba(X_test[self.features_important_linear])[:, 1]
        return prob_train_1*0.5 + prob_train_2*0.33 + prob_train_2*0.17
    
    def predict(self, df_contract, df_personal, df_internet, df_phone, threshold = 0.5):
        return self.predict_proba(df_contract, df_personal, df_internet, df_phone) > threshold
```
