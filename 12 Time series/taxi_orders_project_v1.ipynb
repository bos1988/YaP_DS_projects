#  Прогнозирование количества заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Вам нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

## Цель исследования

- Создать модель для прогнозирования количества заказов такси в аэропортах на следующий час.
- Добится показатей метрики RMSE менее 48.

```python
# Рабочая версия с обработкой Pipeline
# Проходит не по всей сетке, а пока растет значение метрики (Бейесовская оптимизация).
# Если метрика не растет на протажении хвоста (max_tail) цикл заканчивается
# На вход подается Класс модели, исследуемые гиперапараметры в виде словаря
# Работает быстрее, чем GridSearchCV
def ascending_grid(df_features, df_target, func, parameters, scoring=None, plot=False, subtitle='', max_tail=None, cv=5, **kwargs):
    #cv = 5 #при значении 5 результаты отличаются от результатов при значении > 10 ! но по умолчанию в cross_val_score стоит 5, поэтому оставим по умоланию 5
    steps = 100
    iterator = 0
    best_score = -1_000_000
    best_param = parameters
    tail = 0
    if max_tail == None:
        max_tail = 1 if 5 - len(parameters) < 1 else 5 - len(parameters)
    
    param = list(parameters.keys())[0]
    start = parameters[param][0]
    sep = parameters[param][1]
    
    if len(parameters) == 1:
        for p in np.arange(start, start+steps*sep, sep):
            set_param = {param: p}
            if type(func) == Pipeline:
                model = func
                model_name = model.get_params().get("steps")[-1][0]
                # Гиперпараметры (стандартные random_state=123, n_jobs=n_jobs задаются извне):
                # Текущая настройка
                model.set_params(**{model_name+'__'+param: p})
                # **kwargs
                params_temp = {}
                for key, val in kwargs.items():
                    params_temp[model_name+'__'+key] = val
                model.set_params(**params_temp)
            else:
                model = func(random_state=123, **set_param, **kwargs)
#                 model = func(random_state=123, n_jobs=n_jobs, **set_param, **kwargs)

            scores = cross_val_score(model, df_features, df_target, scoring=scoring, cv=cv)
            score = scores.mean()
            iterator += len(scores)
            if plot:
                print(f'{p} >>> {set_param} {kwargs} - {score:.4f}')

            if score > best_score:
                best_score = score
                set_param.update(kwargs)
                best_param = set_param
                tail = 0
            else:
                tail += 1
            if tail >= max_tail:
                break
    else:
        # рекурсия
        parameters_minus = parameters.copy()
        del parameters_minus[param]
        for p in np.arange(start, start+steps*sep, sep):
            set_param = {param: p}
            result = ascending_grid(df_features, df_target, func, parameters_minus,
                                    scoring=scoring, plot=plot, subtitle=subtitle, max_tail=max_tail, cv=cv,
                                    **set_param, **kwargs)
            iterator += result['iterations']
            score = result['score']
            
            # Сокращение нижней границы поиска (предыдущее лучшее минус 2 промежутка)
            for key in result['parameters'].keys():
                try:
                    new_start = result['parameters'][key] - parameters_minus[key][1]*2
                    if parameters_minus[key][0] < new_start:
                        parameters_minus[key][0] = result['parameters'][key] - parameters_minus[key][1] * (max_tail + 1)
                except:
                    pass
            
            if score > best_score:
                best_score = score
                best_param = result['parameters']
                tail = 0
            else:
                tail += 1
            if tail >= max_tail:
                break
        
        if plot:
            print('best_score =', best_score)
            print('best_param =', best_param)
    if plot:
        print('-'*TABLE_SIZE)
    return ({'iterations': iterator, 'score':best_score, 'parameters':best_param})
```

```python
# Кросс-валидация с результатом для одной модели
def cross_val_base(model_name, model, X, y, scoring, print_res=True):
    time_start = time.time()
    scores = cross_val_score(model, X, y, scoring=scoring, cv=10)
    if print_res:
        print(f'Среднее значение метрики {scoring} для базовой модели {text.BOLD}{model_name}:\n{text.BG_SILVER} {scoring} = {scores.mean():.2f} {text.END}')
        print('-'*TABLE_SIZE)
        return
    return (model_name, scores.mean(), time.time() - time_start)
```

```python
# вывод таблицы сравнения
def get_selection_model_results(results, score_name = 'Score'):
    res = pd.DataFrame(results, columns = ['Model', score_name, 'Time']).sort_values(by=score_name, ascending=False)
    
    res['rating_score'] = res.apply(
        lambda row: round((row[score_name] - res[score_name].min()) / (res[score_name].max() - res[score_name].min()), 3)
        , axis=1)
    res['rating_time'] = res.apply(
        lambda row: round(-(row['Time'] - res['Time'].min()) / (res['Time'].max() - res['Time'].min()), 3)
        , axis=1)
    res['rating_total'] = res['rating_score'] + res['rating_time']
    
    return res
```

## Подготовка данных

### Функция подготовки данных

Создадим функцию создания и подготовки основного набора признаков (для последующей оценки важности признаков):
1. произведем ресемплинг данных
2. создадим календарные признаки, ктороые могут оказать влияние на предсказания:
    - час (ввиду суточной сезонности)
    - день недели (ввиду недельнйо сезонности)
    - день (нуждаестя в проверке ввиду наличия цикличных гребней на дневных графиках тренда)
3. создадим признаки "отстающих значений"
4. создадим признаки скользящих: средних, стандартных отклонений, минимального, максимального
5. удалим пропуски по вновь созданным признакам

\* *(год не рассматриваем, потому что располагаем данными за один год)*<br/>
\* *(месяц не рассматриваем, потому что располагаем данными за период менее года и нет возможности оценить месячную сезонность и горизонт прогнозирования довольно мал для влияния месяца)*

```python
# генерация признаков для временного ряда
def make_df_major(df_inp, col, calendar_list, lag_list,
                  rolling_list,
                  resampling='1H'):
    # Ресемплинг
    df = df_inp.resample(resampling).sum()
    # Календарные признаки
    for p in calendar_list:
        df[p] = eval(f'df.index.{p}')
    # «Отстающие значения»
    for lag in lag_list:
        df[f'lag_{lag}'] = df[col].shift(lag)
    # Скользящее
    for roll in rolling_list:
        df[f'r_mean_{roll}'] = df[col].shift().rolling(roll).mean()
        df[f'r_std_{roll}'] = df[col].shift().rolling(roll).std()
        df[f'r_min_{roll}'] = df[col].shift().rolling(roll).min()
        df[f'r_max_{roll}'] = df[col].shift().rolling(roll).max()
    
    df = df.dropna()
    
    return df
```

```python
# Создадим функцию кросс-валидации для временных рядов
# Обучается каждый раз на наборе равном TRAIN_PERIOD. Валидируется по размеру cv
# Вследствие подгонки функции под стандарный вид (для возможности использования в других функциях) не могу добавить TRAIN_PERIOD еще одним параметром, поэтому взял глобальную переменную (что, понимаю, не хорошо)
def cross_val_score(estimator, X, y=None, *, scoring=None, cv=5, n_jobs=-1):
    res = []
    step = (len(X) - TRAIN_PERIOD) // cv
    for i in range(cv):
        index_start = i * step
        index_end = index_start + TRAIN_PERIOD
        
        X_train = X.iloc[index_start:index_end] # обучающая = длине TRAIN_PERIOD
        X_valid = X.iloc[index_end:index_end+step] # валидная = (длина(X) - TRAIN_PERIOD) / cv
        y_train = y[X_train.index]
        y_valid = y[X_valid.index]
                
        estimator.fit(X_train, y_train)
        scorer = get_scorer(scoring)
        
        res.append(scorer(estimator, X_valid, y_valid))
    return np.array(res)
```

### Основной рабочий код

#### Рабочие функции:

```python
def train_preparation(df_inp):
    df = df_inp.copy()
    
    # Настройки подготовки
    target = 'num_orders'
    calendar_list = ['hour']
    lag_list = [168, 48, 24, 2, 1]
    rolling_list = rolling_list = range(7, 7*4+1, 7)
    
    # Подготовка (генерация признаков)
    # Календарные признаки
    for p in calendar_list:
        df[p] = eval(f'df.index.{p}')
    # «Отстающие значения»
    for lag in lag_list:
        df[f'lag_{lag}'] = df[target].shift(lag)
    # Скользящее
    for roll in rolling_list:
#         df[f'r_mean_{roll}'] = df[col].shift().rolling(roll).mean()
        df[f'r_std_{roll}'] = df[target].shift().rolling(roll).std()
        df[f'r_min_{roll}'] = df[target].shift().rolling(roll).min()
        df[f'r_max_{roll}'] = df[target].shift().rolling(roll).max()
    
    df = df.dropna()
    
    return df
```

```python
def get_model(df_train):
    # обработка данных:
    df = train_preparation(df_train)
    X_train, y_train = xy_split(df, 'num_orders')
    
    # Подбор гиперапаратетров, обучение:
    parameters = {'alpha': [0.1, 0.1]}
    best_params = ascending_grid(X_train, y_train,
                                 Lasso,
                                 parameters, scoring='neg_root_mean_squared_error', plot=False, cv=10)['parameters']
    
    model = Lasso(**best_params, random_state=123)
    model.fit(X_train[-1464:], y_train[-1464:])
    
    return model
```

```python
def get_one_prediction(datetime, work_model, database):
    # Настройки подготовки
    target = 'num_orders'
    lag_list = {168: datetime-pd.Timedelta(hours=168),
                48: datetime-pd.Timedelta(hours=48),
                24: datetime-pd.Timedelta(hours=24),
                2: datetime-pd.Timedelta(hours=2),
                1: datetime-pd.Timedelta(hours=1)
               }
    rolling_list = range(7, 7*4+1, 7)
    
    # Возьмем уже известные значения из имеющейся базы данных и создадим на их основе признаки для предсказаний
    df = database.loc[datetime-pd.Timedelta(hours=200):datetime-pd.Timedelta(hours=1)]
    results = {}
    
    # Календарные признаки
    results['hour'] = datetime.hour
    # «Отстающие значения»
    for lag, index in lag_list.items():
        results[f'lag_{lag}'] = df.loc[index, target]
    # Скользящее
    for roll in rolling_list:
#         df[f'r_mean_{roll}'] = df[col].rolling(roll).mean()
        results[f'r_std_{roll}'] = df[target].rolling(roll).std().iloc[-1]
        results[f'r_min_{roll}'] = df[target].rolling(roll).min().iloc[-1]
        results[f'r_max_{roll}'] = df[target].rolling(roll).max().iloc[-1]
    
    return work_model.predict(pd.DataFrame([results])).round()
```

```python
def get_predictions(date_list, work_model, database):
    return [get_one_prediction(row, work_model, database)[0] for row in date_list]
```
