### Функции машинного обучения

```python
# ФУНКЦИИ:
target_col = 'exited'
required_score = 0.59

#=====================================================================================================================
# Машинное обучение:

# Функции балансировки
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)
    
    return features_downsampled, target_downsampled


# *****************************************************************************

# Подготовка данных для обучения
# df на входе должен быть OHE для scaling
# >> df_ml список списков [features/target][train/valid]
def make_ml_data(df, target=target_col, random_state=None, subset=['']):
    df_features = df.drop(target, axis=1)
    df_target = df[target]
    features_train, features_valid, target_train, target_valid = train_test_split(df_features, df_target, test_size=0.25, random_state=random_state)
    df_ml = {
        'features': {
            'train': features_train,
            'valid': features_valid,
        },
        'target': {
            'train': target_train,
            'valid': target_valid,
        }
    }
    
    if type(subset) == str:
        subset = [subset]

    if any(s.startswith('scale') for s in subset):
        scaler = StandardScaler()
        numeric = [col for col in df.columns if df[col].nunique() > 2] #численные колонки (остальные: 0 и 1 - 2 значения)
        scaler.fit(df_ml['features']['train'][numeric])
        df_ml['features']['train'][numeric] = scaler.transform(df_ml['features']['train'][numeric])
        df_ml['features']['valid'][numeric] = scaler.transform(df_ml['features']['valid'][numeric])
    
    if any(s.startswith('up') for s in subset):
        ratio = df_ml['target']['train'].value_counts(normalize=True)
        sampling_mult = round(ratio[0] / ratio[1])
        df_ml['features']['train'], df_ml['target']['train'] = upsample(df_ml['features']['train'], df_ml['target']['train'], sampling_mult)
    if any(s.startswith('down') for s in subset):
        ratio = df_ml['target']['train'].value_counts(normalize=True)
        sampling_mult = round(ratio[1] / ratio[0], 2)
        df_ml['features']['train'], df_ml['target']['train'] = downsample(df_ml['features']['train'], df_ml['target']['train'], sampling_mult)

    return df_ml


# Получить ОДНУ оценку для деревянной модели
def get_tree_score(df_ml, model_func, metric, **kwargs):
    model = model_func(**kwargs, random_state=123)
    model.fit(df_ml['features']['train'], df_ml['target']['train'])
    predictions = model.predict(df_ml['features']['valid'])
    return metrics[metric](df_ml['target']['valid'], predictions)

# Получить ОДНУ / СЛОВАРЬ оценок для модели регрессии (если threshold задан, то - одно значение, если не задан - словарь оценок для 20-ти порогов)
def get_regression_score(df_ml, metric, threshold=None, **kwargs):
    model = LogisticRegression(**kwargs, random_state=123)
    model.fit(df_ml['features']['train'], df_ml['target']['train'])
    probabilities_valid = model.predict_proba(df_ml['features']['valid'])
    probabilities_one_valid = probabilities_valid[:, 1]
    
    if threshold == None:
        results = {}
        for th in np.arange(0, 1.05, 0.05):
            predictions = probabilities_one_valid > th
            results[round(th, 2)] = metrics[metric](df_ml['target']['valid'], predictions)
        return results
    else:
        predictions = probabilities_one_valid > threshold
        return metrics[metric](df_ml['target']['valid'], predictions)


# Набор score для набора гиперапараметров для ЗАДАННОЙ МОДЕЛИ (случайная разбивка данных):
def random_results_for_one_tree(df, model_func, metric, target=target_col, count=10, subset=[''], **kwargs):
    results = []
    for _ in range(count):
        df_ml = make_ml_data(df, subset=subset)
        results.append(get_tree_score(df_ml, model_func, metric, **kwargs))
    return pd.DataFrame(results, columns=[metric])

def random_results_for_one_regression(df, metric, threshold, count=10, subset=[''], **kwargs):
    results = []
    if type(subset) == str:
        subset = [subset]
    subset.append('scale')
    for _ in range(count):
        df_ml = make_ml_data(df, subset=subset)
        results.append(get_regression_score(df_ml, metric, threshold=threshold, **kwargs))
    return pd.DataFrame(results, columns=[metric])


# Распределение для набора результатов случаных разбивок и одной группы гиперпараметров
def show_results_for_one_param(df_simple, metric, model_name='МОДЕЛЬ', subtitle='', distr=True): #subtitle=f'\nпри max_depth={max_depth}
    hist_box_describe(df_simple,
                      metric,
                      subtitle=f'\n{subtitle} для cлучайных разбивок данных\n(набор из {df_simple.shape[0]} значений)',
                      xlabel=f'{metric} модели {model_name}',
                      bins=100)
    if distr:
        accuracy_distribution(df_simple, metric)

def show_results_for_one_tree_2(df, metric, max_depth=1, count=10, figsize=(8, 8), distr=True, **kwargs):
    show_results_for_one_param(random_results_for_one_tree(df,
                                              DecisionTreeClassifier,
                                              metric,
                                              count=count,
                                              max_depth=max_depth, **kwargs), 
                               metric,
                               model_name='ДЕРЕВО РЕШЕНИЙ',
                               subtitle=f'при max_depth={max_depth}', distr=distr)

def show_results_for_one_forest_2(df, metric, max_depth=1, n_estimators=1, count=10, figsize=(8, 8), distr=True, **kwargs):
    show_results_for_one_param(random_results_for_one_tree(df,
                                              RandomForestClassifier,
                                              metric,
                                              count=count,
                                              max_depth=max_depth, n_estimators=n_estimators, **kwargs), 
                               metric,
                               model_name='СЛУЧАЙНЫЙ ЛЕС',
                               subtitle=f'при max_depth={max_depth} и n_estimators={n_estimators}', distr=distr)


# Набор score для заданного диапазона max_depth для ДЕРЕВА РЕШЕНИЙ (случайная разбивка данных):
def results_for_various_tree_2(df, metric, depths=(1, 10, 1, 1), count=10, subset=[''], **kwargs):
    results = {}
    start, end, sep, mult = depths
    for _ in range(count):
        df_ml = make_ml_data(df, subset=subset)
        depth = start
        i = sep
        while(depth < end + 1):
            result = get_tree_score(df_ml, DecisionTreeClassifier, metric, max_depth=depth, **kwargs)
            results.setdefault(depth, []).append(result)
            depth = depth + i
            i += i * (mult - 1)
    return pd.DataFrame(results)

# Набор score для заданного диапазона max_depth и n_estimators для СЛУЧАНОГО ЛЕСА (случайная разбивка данных):
def results_for_various_forest_2(df, metric, depths=(1, 5, 1, 1), n_estimators_count=(1, 5, 1, 1), count=10, subset=[''], **kwargs):
    results = {'max_depth': [], 'n_estimators': [], metric: []}
    start_d, end_d, sep_d, mult_d = depths
    start_e, end_e, sep_e, mult_e = n_estimators_count

    for _ in range(count):
        df_ml = make_ml_data(df, subset=subset)
        depth = start_d
        i = sep_d
        while(depth < end_d + 1):
#             print('1 start_d:', depth, '/', 'end_d:', end_d)
            estim = start_e
            j = sep_e
            while(estim < end_e + 1):
#                 print('        1 start_e:', estim, '/', 'end_e:', end_e)
                result = get_tree_score(df_ml, RandomForestClassifier, metric, max_depth=depth, n_estimators=estim, **kwargs)
                results['max_depth'].append(depth)
                results['n_estimators'].append(estim)
                results[metric].append(result)
                estim = estim + j
                j += j * (mult_e - 1)
#                 print('        2 start_e:', estim, '/', 'end_e:', end_e)
            depth = depth + i
            i += i * (mult_d - 1)
#             print('2 start_d:', depth, '/', 'end_d:', end_d)

    return pd.DataFrame(results)

# Набор score для различных порогов для ЛОГИСТИЧЕСКОЙ РЕГРЕССИИ (случайная разбивка данных):
def results_for_various_regression_2(df, metric, count=10, subset=[''], **kwargs):
    results = []
    if type(subset) == str:
        subset = [subset]
    subset.append('scale')
    for _ in range(count):
        df_ml = make_ml_data(df, subset=subset)
        results.append(get_regression_score(df_ml, metric, **kwargs))

    return pd.DataFrame(results)


# Вывод boxplot-ов для набора score для различных параметров модели (данные на вход - сформированные)
def show_results_on_boxplot(data, metric, model_name='МОДЕЛЬ', params=['параметров'], figsize=(8, 8)):
    plt.figure(figsize=(figsize[0], figsize[1]))
    sns.boxplot(data=data)
    plt.grid(visible=True, axis='both')
    plt.title(f'Распределения значений\n{metric} модели {model_name}\nдля различных {",".join(params)}\nдля cлучайных разбивок данных\n(наборы из {data.shape[0]} значений)', pad=15, fontsize = 14)
    plt.xlabel(f'Значение {",".join(params)}', fontsize = 14)
    plt.ylabel(f'Значение {metric}', fontsize = 14)
    plt.show()

# Вывод boxplot-ов для набора score для различных max_depth для ДЕРЕВА РЕШЕНИЙ (обычно по результатам results_for_various_tree())
def show_results_for_various_tree(data, metric, figsize=(8, 8)):
    show_results_on_boxplot(data, metric, model_name='ДЕРЕВО РЕШЕНИЙ', params=['MAX_DEPTH'], figsize=(8, 8))

# Вывод boxplot-ов для набора score для различных порогов для ЛОГИСТИЧЕСКОЙ РЕГРЕССИИ (обычно по результатам !!!!!! results_for_various_regression())
def show_results_for_various_regression(data, metric, figsize=(8, 8)):
    show_results_on_boxplot(data, metric, model_name='ЛОГИСТИЧЕСКАЯ РЕГРЕССИЯ', params=['THRESHOLD'], figsize=(8, 8))

# Вывод heatmap для набора score для различных max_depth и n_estimators для СЛУЧАНОГО ЛЕСА (обычно по результатам results_for_various_forest())
def show_results_for_various_forest(data, metric, figsize=(8, 8)):
    plt.figure(figsize=(figsize[0], figsize[1]))
    sns.heatmap(data.pivot_table(index='n_estimators', columns='max_depth', values=metric, aggfunc='median'),
                cmap='Purples', annot=True, fmt='.3f')
    plt.title(f'Медианные значения {metric} модели СЛУЧАЙНЫЙ ЛЕС\nв зависимости от MAX_DEPTH и N_ESTIMATORS\nдля cлучайных разбивок данных\n(наборы из {data.groupby(["max_depth", "n_estimators"]).count().max().values[0]} значений)', pad=15, fontsize = 14)
    plt.xlabel('Значение max_depth', fontsize = 14)
    plt.ylabel('Значение n_estimators', fontsize = 14)
    plt.yticks(rotation=0)
    plt.tick_params(labelsize = min(figsize)*1.5)
    plt.show()


# Распределение accuracy и вероятность выпадения менее required_score при случайном построении модели
def accuracy_distribution(data, metric, figsize=(10, 8)):
    # Нормальное распределение
    min_value = data.min().values[0]
    max_value = data.max().values[0]
    
    mu = data.mean().values[0]
    sigma = np.std(data, ddof=1).values[0]

    x = [i/1000 for i in range(int(min_value*1000-20), int(max_value*1000+20))]
    distr = st.norm(mu, sigma)

    fig, ax1 = plt.subplots()
    ax2 = ax1.twinx()
    ax1.plot(x, distr.pdf(x), label = 'Нормальное \nраспределние', color='r', linewidth=4)
    ax2.plot(x, distr.cdf(x), label = 'Вероятности', color='b', linestyle='--', linewidth=2)
    ax2.plot([required_score, required_score], [0, 1], label = 'Искомое \nзначение', linestyle=':', color='g', linewidth=4)

    fig.set_figwidth(figsize[0])
    fig.set_figheight(figsize[1])

    #ax1.xaxis.tick_top()
    ax1.minorticks_on()
    ax1.grid(which='major',
            color = 'grey', 
            linewidth = 1.2)
    ax1.grid(which='minor', 
            color = 'grey', 
            linestyle = ':')

    ax1.legend(fontsize = 12, loc = 'center left')
    ax2.legend(fontsize = 12, loc = 'center right')


    ax1.set_xlabel(f'Значения {metric}', fontsize = 14)
    ax1.set_ylabel('Частота для нормального распределения', fontsize = 14)
    ax2.set_ylabel('Вероятность значения больше данного', fontsize = 14)
    ax1.tick_params(labelsize = 12)
    ax2.tick_params(labelsize = 12)
    ax1.set_title(f'Нормальное распределение {metric}\n(набор из {data.shape[0]} значений)', pad=15, fontsize = 16)

    ax2.text(min_value-0.02, 0.9, f'Вероятность меньше {required_score}:\n{distr.cdf(required_score):.3%}',
            rotation = 0,
            fontsize = min(figsize)*1.75,
            bbox = {'facecolor':'white', 'pad': min(figsize)}
                )
    ax2.text(min_value-0.02, 0.3, f'Среднее = {mu:.4f}\nОтклонение = {sigma:.4f}',
            rotation = 0,
            fontsize = min(figsize)*1.5,
            bbox = {'facecolor':'white', 'pad': min(figsize)}
                )
    plt.show()
    print('-'*table_size)
    print(f'Вероятность получить значение меньше {required_score} = {distr.cdf(required_score):.3%}')
    print('-'*table_size)
    print()

# Отрисовка кривой roc_auc
def print_roc_curve(model, features, target, subtitle='', figsize=(7, 7)):
    probabilities_test = model.predict_proba(features)
    probabilities_one_test = probabilities_test[:, 1]
    roc_auc = roc_auc_score(target, probabilities_one_test)
    
    fpr, tpr, thresholds = roc_curve(target, probabilities_one_test) 

    plt.figure(figsize=(figsize[0], figsize[1]))
    plt.plot(fpr, tpr, linewidth=3, color = 'darkcyan')
    plt.fill_between(fpr, tpr, 0, color = 'darkcyan', alpha=0.15)

    # ROC-кривая случайной модели (выглядит как прямая)
    plt.plot([0, 1], [0, 1], linestyle='--', color = 'darkkhaki')

    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel("False Positive Rate", fontsize = 14)
    plt.ylabel("True Positive Rate", fontsize = 14)
    plt.title(f"ROC-кривая{subtitle}", pad=15, fontsize = 16)


    plt.text(0.4, 0.1, f'Метрика roc_auc:\n{roc_auc:.4f}',
                rotation = 0,
                fontsize = min(figsize)*1.8,
                bbox = {'facecolor':'white', 'boxstyle': 'round', 'pad': min(figsize)/12}
                    )
    plt.show()
```

### Функции

#### Универсальная функция поиска ОДНОГО гиперпараметра

```python
# Проходит не по всей сетке, а пока растет значение метрики.
# Если метрика не растет на протажении хвоста (max_tail) цикл заканчивается
# На вход подается Класс модели, название исследуемого гиперапараметра
def get_best_param(df_features, df_target, func, param, ranging=range(1, 51, 1), scoring=None, plot=True, subtitle='', max_tail=4, **kwargs):
    r_cv = (10 if len(df_target) > 1000 else 5)
    best_param = ranging[0]
    best_score = 0
    tail = 0
    results = []
    x = []
    y = []
    for p in ranging:
        set_param = {param: p}
        model = func(random_state=123, **set_param, **kwargs)
        scores = cross_val_score(model, df_features, df_target, scoring=scoring, cv=r_cv)
        score = scores.mean()
        x.append(p)
        y.append(score)
        if score > best_score:
            best_score = score
            best_param = p
            tail = 0
        else:
            tail += 1
        if tail >= max_tail:
            break
    if plot:
        plt.figure(figsize=(6, 4))
        plt.plot(x, y, linewidth=3, color = 'darkcyan')
        plt.grid(visible=True, axis='both')
        func_name = str(func).split('.')[-1][:-2]
        plt.title(f'Средние значения метрики {scoring}\nмодели {func_name}\nдля различных {param} {subtitle}\n(число валидационных разбивок cv = {r_cv})', pad=15, fontsize = 14)
        plt.xlabel(f'Значение {param}', fontsize = 14)
        plt.ylabel(f'Значение {scoring}', fontsize = 14)
        plt.show()
        print(f'Лучший {text.BOLD}{param}{text.END} = {text.BG_YELLOW} {best_param} {text.END} с {text.BOLD}{scoring}{text.END} = {text.BG_SILVER} {best_score:.4f} {text.END}')
        print('-'*table_size, end='\n')
    
    return (best_param, best_score)
```

#### СУПЕР функция вместо GridSearchCV

```python
# Рабочая версия
# Проходит не по всей сетке, а пока растет значение метрики (Бейесовская оптимизация).
# Если метрика не растет на протажении хвоста (max_tail) цикл заканчивается
# На вход подается Класс модели, исследуемые гиперапараметры в виде словаря
# Работает быстрее, чем GridSearchCV
def ascending_grid(df_features, df_target, func, parameters, scoring=None, plot=False, subtitle='', max_tail=None, cv=5, **kwargs):
    #cv = 5 #при значении 5 результаты отличаются от результатов при значении > 10 ! но по умолчанию в cross_val_score стоит 5, поэтому оставим по умоланию 5
    steps = 100
    iterator = 0
    best_score = 0
    best_param = parameters
    tail = 0
    if max_tail == None:
        max_tail = 1 if 5 - len(parameters) < 1 else 5 - len(parameters)
    
    param = list(parameters.keys())[0]
    start = parameters[param][0]
    sep = parameters[param][1]
    
    if len(parameters) == 1:
        for p in range(start, start+steps*sep, sep):
            set_param = {param: p}
            model = func(random_state=123, **set_param, **kwargs)
            scores = cross_val_score(model, df_features, df_target, scoring=scoring, cv=cv)
            score = scores.mean()
            iterator += len(scores)
            if plot:
                print(f'{p} >>> {set_param} {kwargs} - {score:.4f}')

            if score > best_score:
                best_score = score
                set_param.update(kwargs)
                best_param = set_param
                tail = 0
            else:
                tail += 1
            if tail >= max_tail:
                break
    else:
        # рекурсия
        parameters_minus = parameters.copy()
        del parameters_minus[param]
        for p in range(start, start+steps*sep, sep):
            set_param = {param: p}
            result = ascending_grid(df_features, df_target, func, parameters_minus,
                                    scoring=scoring, plot=plot, subtitle=subtitle, max_tail=max_tail, cv=cv,
                                    **set_param, **kwargs)
            iterator += result['iterations']
            score = result['score']
            
            # Сокращение нижней границы поиска (предыдущее лучшее минус 2 промежутка)
            for key in result['parameters'].keys():
                try:
                    new_start = result['parameters'][key] - parameters_minus[key][1]*2
                    if parameters_minus[key][0] < new_start:
                        parameters_minus[key][0] = result['parameters'][key] - parameters_minus[key][1] * (max_tail + 1)
                except:
                    pass
            
            if score > best_score:
                best_score = score
                best_param = result['parameters']
                tail = 0
            else:
                tail += 1
            if tail >= max_tail:
                break
        
        if plot:
            print('best_score =', best_score)
            print('best_param =', best_param)
    if plot:
        print('-'*table_size)
    return ({'iterations': iterator, 'score':best_score, 'parameters':best_param})
```

```python
# На всход подается созданная необученная модель (не класс)
def get_best_threshold(model, df_features, df_target, metric, plot=True, subtitle='', sep=0.05):
    if metric == 'roc_auc':
        print('Special roc_auc')
        return None
    r_cv = (10 if len(df_target) > 1000 else 5)
    best_th = 0
    best_score = 0
    results = []
    span = len(df_target) // r_cv

    results = []
    for th in np.arange(0, 1.05, sep):
        scores = []
        for i in range(0, len(df_target), span):
            Y_valid = df_target[i: i+span]
            Y_train = df_target.drop(Y_valid.index)
            X_valid = df_features.loc[Y_valid.index]
            X_train = df_features.loc[Y_train.index]
            
            # отключать ли масштабирование? И проводить ли на деревьях поиск порога, там нет эффекта и пороги там другие
            scaler = StandardScaler()
            numeric = [col for col in df_features.columns if df_features[col].nunique() > 2] #численные колонки (остальные: 0 и 1 - 2 значения)
            scaler.fit(X_train[numeric])
            X_train[numeric] = scaler.transform(X_train[numeric])
            X_valid[numeric] = scaler.transform(X_valid[numeric])
            
            model.fit(X_train, Y_train)
            probabilities_valid = model.predict_proba(X_valid)
            probabilities_one_valid = probabilities_valid[:, 1]
            
            predictions = probabilities_one_valid > th
            scores.append( metrics[metric](Y_valid, predictions) )
        score = np.mean(scores)
        results.append(score)
        if score > best_score:
            best_score = score
            best_th = th

    if plot:
        plt.figure(figsize=(6, 4))
        plt.plot(np.arange(0, 1.05, sep), results, linewidth=3, color = 'darkcyan')
        plt.grid(visible=True, axis='both')
        plt.title(f'Средние значения метрики {metric}\nдля различных threshold {subtitle}\n(число валидационных разбивок cv = {r_cv})', pad=15, fontsize = 14)
        plt.xlabel(f'Значение threshold', fontsize = 14)
        plt.ylabel(f'Значение {metric}', fontsize = 14)
        plt.show()
        print(f'Лучшее {text.BOLD}threshold{text.END} = {text.BG_YELLOW} {best_th:.3f} {text.END} с {text.BOLD}{metric}{text.END} = {text.BG_SILVER} {best_score:.4f} {text.END}')
        print('-'*table_size, end='\n')
    
    return (round(best_th, 6), best_score)
```
