```python
# выборка колонок по stage, param_type, param_name
# на входе: df и списки (в список можно отрывок названия)
def qdf(df, *args):
    args = list(args[:3])
    names = list(df.columns[1:])
    for i in range(len(args)):
        if type(args[i]) == str:
            args[i] = [args[i]]
        if args[i]:
            names = list(filter(
                    lambda x: any([s in x.split('.')[i] for s in args[i]]), names
                ))
    if not args or 'date' in args[0] or len(args[0]) == 0:
        names.append('date')
    
#     print(len(names))
#     print(*names, sep='\n')
    return df[names] if len(names) > 0 else None
```

### Функции машинного обучения

Поиск гиперпараметров (кросс-валидация)

```python
# Рабочая версия с обработкой Pipeline
# Проходит не по всей сетке, а пока растет значение метрики (Бейесовская оптимизация).
# Если метрика не растет на протажении хвоста (max_tail) цикл заканчивается
# На вход подается Класс модели, исследуемые гиперапараметры в виде словаря
# Работает быстрее, чем GridSearchCV
def ascending_grid(df_features, df_target, func, parameters, scoring=None, plot=False, subtitle='', max_tail=None, cv=5, n_jobs=-1, **kwargs):
    #cv = 5 #при значении 5 результаты отличаются от результатов при значении > 10 ! но по умолчанию в cross_val_score стоит 5, поэтому оставим по умоланию 5
    steps = 100
    iterator = 0
    best_score = -1_000_000
    best_param = parameters
    tail = 0
    if max_tail == None:
        max_tail = 1 if 5 - len(parameters) < 1 else 5 - len(parameters)
    
    param = list(parameters.keys())[0]
    start = parameters[param][0]
    sep = parameters[param][1]
    
    if len(parameters) == 1:
        for p in np.arange(start, start+steps*sep, sep):
            set_param = {param: p}
            if type(func) == Pipeline:
                model = func
                model_name = model.get_params().get("steps")[-1][0]
                # Гиперпараметры (стандартные random_state=123, n_jobs=n_jobs задаются извне):
                # Текущая настройка
                model.set_params(**{model_name+'__'+param: p})
                # **kwargs
                params_temp = {}
                for key, val in kwargs.items():
                    params_temp[model_name+'__'+key] = val
                model.set_params(**params_temp)
            else:
                model = func(random_state=123, n_jobs=n_jobs, **set_param, **kwargs)
            scores = cross_val_score(model, df_features, df_target, scoring=scoring, cv=cv)
            score = scores.mean()
            iterator += len(scores)
            if plot:
                print(f'{p} >>> {set_param} {kwargs} - {score:.4f}')

            if score > best_score:
                best_score = score
                set_param.update(kwargs)
                best_param = set_param
                tail = 0
            else:
                tail += 1
            if tail >= max_tail:
                break
    else:
        # рекурсия
        parameters_minus = parameters.copy()
        del parameters_minus[param]
        for p in np.arange(start, start+steps*sep, sep):
            set_param = {param: p}
            result = ascending_grid(df_features, df_target, func, parameters_minus,
                                    scoring=scoring, plot=plot, subtitle=subtitle, max_tail=max_tail, cv=cv,
                                    **set_param, **kwargs)
            iterator += result['iterations']
            score = result['score']
            
            # Сокращение нижней границы поиска (предыдущее лучшее минус 2 промежутка)
            for key in result['parameters'].keys():
                try:
                    new_start = result['parameters'][key] - parameters_minus[key][1]*2
                    if parameters_minus[key][0] < new_start:
                        parameters_minus[key][0] = result['parameters'][key] - parameters_minus[key][1] * (max_tail + 1)
                except:
                    pass
            
            if score > best_score:
                best_score = score
                best_param = result['parameters']
                tail = 0
            else:
                tail += 1
            if tail >= max_tail:
                break
        
        if plot:
            print('best_score =', best_score)
            print('best_param =', best_param)
    if plot:
        print('-'*TABLE_SIZE)
    return ({'iterations': iterator, 'score':best_score, 'parameters':best_param})
```

```python
def changes_by_stages(df, metal):
    temp = {}
    for col in [f'concentrate_{metal}', f'tail_{metal}']:
        temp[col] = {}
        for row in ['rougher', 'primary', 'secondary', 'final']:
            select = qdf(df, row, '', [col])
            try:
                temp[col][row] = select.mean()[0]
            except:
                temp[col][row] = '-'
    return pd.DataFrame(temp)
```

#### Эффективность обогащения чернового концентрата 

Вычислим эффективность обогащения на полной выборке для признака rougher.output.recovery.

```python
# Функция вычисления эффективности обогащения
def get_recovery(df, *, C=None, F=None, T=None):
    numerator = df[C] * (df[F] - df[T])
    denominator = df[F] * (df[C] - df[T])
    return numerator / denominator * 100
```

Исходный целевой признак (эффективность обогащения):

```python
df_prep[target_col_1].head()
```

Рассчитанная эффективность обогащения:

```python
# доля золота в концентрате после флотации:
coef_C = 'rougher.output.concentrate_au'
# доля золота в сырье до флотации:
coef_F = 'rougher.input.feed_au'
# доля золота в отвальных хвостах после флотации:
coef_T = 'rougher.output.tail_au'
get_recovery(df_prep, C=coef_C, F=coef_F, T=coef_T).head()
```

Найдем MAE между расчётами и значением признака.

```python
round(mean_squared_error(df_prep[target_col_1], get_recovery(df_prep, C=coef_C, F=coef_F, T=coef_T)), 10)
```

Проверим гипотезу:
> Конентрация золота увеличивается в ходе процесса очистки руды.

Нулевая гитотеза: Конентрация золота не изменяется в ходе процесса очистки руды.

Альтернативная гипотеза: Конентрация золота увеличивается в ходе процесса очистки руды.

Воспользуемся методом проверки односторонней гипотезы.

Критический уровень значимости примем 1%

```python
rougher_to_final = df_prep[f'final.output.concentrate_{element}'] / df_prep[f'rougher.output.concentrate_{element}']
rougher_to_final = rougher_to_final.dropna()

interested_value = 1
alpha = 0.01

results = st.ttest_1samp(rougher_to_final, interested_value)

print('p-значение:', results.pvalue)

if (results.pvalue / 2 < alpha and rougher_to_final.mean() > interested_value):
    print(f"Отвергаем нулевую гипотезу: концентрация {element_name} увеличивается")
else:
    print(f"Не получилось отвергнуть нулевую гипотезу: концентрация {element_name} не увеличивается")
```

### Функции вычисления метрик

```python
# метрика sMAPE (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»).
# возвращает не процент, а долю от среднего (так удобнее). Абсолютная точность = 0.0
def smape_score(target, predictions):
    if not len(target) == len(predictions):
        print('Длины входных данных не совпадают!')
        return 'error'
    return sum(map(lambda x: 0 if x[0] + x[1] == 0 else abs(x[0]-x[1]) / (abs(x[0]) + abs(x[1])), zip(target, predictions))) / len(target) * 2
    #return sum([abs(t-p) / (abs(t) + abs(p)) for t, p in zip(target, predictions)]) / len(target) * 2

# score will negate the return value of smape_score
# https://scikit-learn.org/stable/modules/model_evaluation.html#scoring
scoring = make_scorer(smape_score, greater_is_better=False)
```

```python
# итоговая метрика
def total_smape(rougher_smape, final_smape):
    return 0.25 * rougher_smape + 0.75 * final_smape
```

```python
# Функция вычисления эффективности обогащения
def get_recovery_array(C=None, F=None, T=None):
    result = []
    for i in range(len(C)):
        numerator = C[i] * (F[i] - T[i])
        denominator = F[i] * (C[i] - T[i])
        result.append(numerator / denominator * 100)
    return result
```
